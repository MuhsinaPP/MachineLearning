# **Cross validation**

```python
x=data.drop('diagnosis', axis=1)
y=data['diagnosis']
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
```

```python
param={'DEcisionTree':DecisionTreeClassifier(),
       'RandomForest':RandomForestClassifier(),
       'SVC':SVC(),
       'KNeighbors':KNeighborsClassifier(),
       'GaussianNB':GaussianNB(),
       'Logistic':LogisticRegression()}
```

```python
cv_result={}
```

```python
for name, clf in param.items():
  scores=cross_val_score(clf,x,y,cv=5)
  cv_result[name]=scores.mean()
```

```python
for name,score in cv_result.items():
  print(f"{name}:{score}")
```

```python
best_param=max(cv_result,key=cv_result.get)
print(best_param,":",cv_result[best_param])
```

# **Split data into training and test sets**

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2, random_state=1)
```

```python
model=RandomForestClassifier()
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

# **Confution matrix**

```python
yact=ytest
ypred=model.predict(xtest)
```
