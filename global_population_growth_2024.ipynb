```python
#Removing the duplicates(if any)

df1=df.drop_duplicates()
df1
```

```python
#Percentage of missing values per column

df1.isna().sum()/df1.shape[0]
```

```python
df1=df1.dropna()
len(df1)
```

```python
# Prepare the data for modeling
x = df[['Population_2023', 'Growth_Rate (%)']]
y = df['Population_2024']
```

```python
sns.regplot(x=df1["Population_2024"],y=y)
```

```python
sns.regplot(x=df1["Growth_Rate (%)"],y=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=42) #result reproducable
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import mean_absolute_error
meanR=mean_absolute_error(y_test,y_pred)
meanR
```

```python
from sklearn.metrics import mean_absolute_percentage_error
perR=mean_absolute_percentage_error(y_test,y_pred)
perR
```

```python
from sklearn.metrics import mean_squared_error
sqrR=mean_squared_error(y_test,y_pred)
sqrR
```

```python
from sklearn.metrics import r2_score
r2R=r2_score(y_test,y_pred)
r2R
```
