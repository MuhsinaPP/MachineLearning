```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/drive/MyDrive/Dataset/heart_2020_cleaned.csv")
```

```python
data
```

```python
data.info()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
```

```python
newHeartDisease=label.fit_transform(data['HeartDisease'])
newSmoking=label.fit_transform(data['Smoking'])
newAlcoholDrinking=label.fit_transform(data['AlcoholDrinking'])
newStroke=label.fit_transform(data['Stroke'])
newDiffWalking=label.fit_transform(data['DiffWalking'])
newSex=label.fit_transform(data['Sex'])
newAgeCategory=label.fit_transform(data['AgeCategory'])
newRace=label.fit_transform(data['Race'])
newDiabetic=label.fit_transform(data['Diabetic'])
newPhysicalActivity=label.fit_transform(data['PhysicalActivity'])
newGenHealth=label.fit_transform(data['GenHealth'])
newAsthma=label.fit_transform(data['Asthma'])
newKidneyDisease=label.fit_transform(data['KidneyDisease'])
newSkinCancer=label.fit_transform(data['SkinCancer'])
```

```python
data['HeartDisease']=newHeartDisease
data['Smoking']=newSmoking
data['AlcoholDrinking']=newAlcoholDrinking
data['Stroke']=newStroke
data['DiffWalking']=newDiffWalking
data['AgeCategory']=newAgeCategory
data['Race']=newRace
data['Diabetic']=newDiabetic
data['PhysicalActivity']=newPhysicalActivity
data['GenHealth']=newGenHealth
data['Asthma']=newAsthma
data['KidneyDisease']=newKidneyDisease
data['SkinCancer']=newSkinCancer
data['Sex']=newSex
```

```python
data.isnull().any()
```

```python
x=data.drop(['HeartDisease'],axis=1)
y=data['HeartDisease']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,random_state=1)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier(n_estimators=50)
```

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
yact=ytest
ypred=model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(yact,ypred)
```

```python
from mlxtend.plotting import plot_confusion_matrix
plot_confusion_matrix(cf)
```
