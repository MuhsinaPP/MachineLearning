**Importing Libraries**

**Loading Dataset**

```python
testdata=pd.read_csv("/content/drive/MyDrive/Dataset/fashion-mnist_test.csv")
traindata=pd.read_csv("/content/drive/MyDrive/Dataset/fashion-mnist_train.csv")
```

**Splitting into train and test datasets**

```python
ytrain=traindata['label']
xtrain=traindata.drop(['label'],axis=1)
```

```python
ytest=testdata['label']
xtest=testdata.drop(['label'],axis=1)
```

**Logistic Regression Classification Model**

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
```

**Fitting the model**

```python
model.fit(xtrain,ytrain)
```

```python
model.score(xtest,ytest)
```

```python
y_act=ytest
y_pred=model.predict(xtest)
```

```python
from sklearn.metrics import confusion_matrix
cf=confusion_matrix(y_act,y_pred)
```

```python
from mlxtend.plotting import plot_confusion_matrix
plot_confusion_matrix(cf)
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_act, y_pred)
print(report)
```
